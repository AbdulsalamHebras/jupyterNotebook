```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
temp_sanaa=[23.1,45.2,45.2,43.6,34.4,45.2,34.2]
temp_ibb=[12.3,23.4,34,56,32,45,23]
```

```python
df_sanaa=pd.DataFrame(data={'Day':days,'temp':temp_sanaa})
df_ibb=pd.DataFrame(data={'Day':days,'temp':temp_ibb})
df_sanaa
```

```python
df_ibb
```

```python
plt.plot(df_sanaa['Day'],df_sanaa['temp'])
plt.show()
```

```python
plt.scatter(df_sanaa['Day'],df_sanaa['temp'])
plt.show()
```

```python
plt.bar(df_sanaa['Day'],df_sanaa['temp'])
plt.show()
```

```python
data=[1,2,0,2,1,4,5,7,2,3,4,2,1,4,4,3,1]
plt.hist(data,bins=5)
plt.show()
```

```python
plt.plot(df_sanaa['Day'],df_sanaa['temp'],label='sanaa',c='r',marker='s',ls='--')
plt.plot(df_ibb['Day'],df_ibb['temp'],label='ibb',c='y', marker='o')
plt.xlabel('days of the week')
plt.ylabel('temp')
plt.title('sanaa')

plt.legend()
plt.show()
```

```python
figure,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].plot(df_sanaa['Day'],df_sanaa['temp'],label='sanaa',c='r',marker='s',ls='--')
axes[0].set_title('sanaa')

axes[1].plot(df_ibb['Day'],df_ibb['temp'],label='ibb',c='y', marker='o')
axes[1].set_title('ibb')
```
